```python
########################################################## Sex and Age #######################################################

import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_excel("dataK.xlsx")


oldM = []
avM = []
youngM = []


oldF = []
avF = []
youngF = []


for index,row in df.iterrows():
    if  20 > row['Age'] and row['Survived'] == 'Yes' and row["Sex"] == 1:
        youngM.append(row['Age'])
    elif row['Age'] > 20 and  40 > row['Age'] and row['Survived'] == 'Yes' and row["Sex"] == 1:
        avM.append(row['Age'])
    elif row["Age"] > 40 and row['Survived'] == 'Yes'and row["Sex"] == 1 and row["Sex"] == 1:
        oldM.append(row['Age'])

for index,row in df.iterrows():
    if  20 > row['Age'] and row['Survived'] == 'Yes' and row["Sex"] == 0:
        youngF.append(row['Age'])
    elif row['Age'] > 20 and  40 > row['Age'] and row['Survived'] == 'Yes' and row["Sex"] == 0:
        avF.append(row['Age'])
    elif row["Age"] > 40 and row['Survived'] == 'Yes'and row["Sex"] == 0 :
        oldF.append(row['Age'])


    
 

data1 = {'Survived Male': [len(oldM),len(avM),len(youngM)]}
df = pd.DataFrame(data1,columns=['Survived Male'],index = ['Senior Adult','Adult','Adolescence '])
df.plot.pie(y='Survived Male',figsize=(8, 8),autopct='%1.1f%%', startangle=90)


data2 = {'Survived Female': [len(oldF),len(avF),len(youngF)]}
df = pd.DataFrame(data2,columns=['Survived Female'],index = ['Senior Adult','Adult','Adolescence '])
df.plot.pie(y='Survived Female',figsize=(8, 8),autopct='%1.1f%%', startangle=90)


plt.show()
```

```python
############################################### Pclass and Embarked #####################################

import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_excel("dataK.xlsx")


Class1EmS = []
Class2EmS = []
Class3EmS = []

Class1EmC = []
Class2EmC = []
Class3EmC = []

Class1EmQ = []
Class2EmQ = []
Class3EmQ = []


for index,row in df.iterrows():
    if  1 == row['Pclass'] and row['Survived'] == 'Yes' and  row["Embarked"] == 'S':
        Class1EmS.append(row['Pclass'])
    elif row['Pclass'] == 2 and row['Survived'] == 'Yes' and row["Embarked"] == 'S':
        Class2EmS.append(row['Pclass'])
    elif row["Pclass"] == 3 and row['Survived'] == 'Yes' and row["Embarked"] == 'S':
        Class3EmS.append(row['Pclass'])

for index,row in df.iterrows():
    if  1 == row['Pclass'] and row['Survived'] == 'Yes' and  row["Embarked"] == 'C':
        Class1EmC.append(row['Pclass'])
    elif row['Pclass'] == 2 and row['Survived'] == 'Yes' and row["Embarked"] == 'C':
        Class2EmC.append(row['Pclass'])
    elif row["Pclass"] == 3 and row['Survived'] == 'Yes' and row["Embarked"] == 'C':
        Class3EmC.append(row['Pclass'])

for index,row in df.iterrows():
    if  1 == row['Pclass'] and row['Survived'] == 'Yes' and  row["Embarked"] == 'Q':
        Class1EmQ.append(row['Pclass'])
    elif row['Pclass'] == 2 and row['Survived'] == 'Yes' and row["Embarked"] == 'Q':
        Class2EmQ.append(row['Pclass'])
    elif row["Pclass"] == 3 and row['Survived'] == 'Yes' and row["Embarked"] == 'Q':
        Class3EmQ.append(row['Pclass'])


data1 = {'Survived': [len(Class1EmS),len(Class2EmS),len(Class3EmS)]}
df = pd.DataFrame(data1,columns=['Survived'],index = ['Class1 Embarked = S','Class2 Embarked = S','Class3 Embarked = S'])
df.plot.pie(y='Survived',figsize=(8, 8),autopct='%1.1f%%', startangle=90)

data2 = {'Survived': [len(Class1EmC),len(Class2EmC),len(Class3EmC)]}
df = pd.DataFrame(data2,columns=['Survived'],index = ['Class1 Embarked = C','Class2 Embarked = C','Class3 Embarked = C'])
df.plot.pie(y='Survived',figsize=(8, 8),autopct='%1.1f%%', startangle=90)

data3 = {'Survived': [len(Class1EmQ),len(Class2EmQ),len(Class3EmQ)]}
df = pd.DataFrame(data3,columns=['Survived'],index = ['Class1 Embarked = Q','Class2 Embarked = Q','Class3 Embarked = Q'])
df.plot.pie(y='Survived',figsize=(8, 8),autopct='%1.1f%%', startangle=90)



plt.show()
```
